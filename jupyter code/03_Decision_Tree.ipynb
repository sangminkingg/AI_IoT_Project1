# 데이터 전처리
- 인코딩 (encoding) : 범주형 데이터를 수치형 데이터로 변화하는 것
  - 인코딩을 하는 이유 : 계산하기 위해
  
  - 라벨(label) 인코딩 : 범주형을 정수로 변환
    - 크기특성이 없는 데이터에 크기를 갖는 숫자를 부여하는 문제
  - one hot encoding : 범주형을 이진수로 변환
    - 크기특성이 없는 데이터에 크기가 없는 숫자를 부여할 수 있음
    - 특성을 늘려주는 효과가 있음
    
- one hot encoding 방법
  - 과목 : 국어, 수학, 과학 -> 국어(001), 수학(010), 과학(100)
  - 해당 값인 경우에만 1이고 그렇지 않으면 0인 컬럼을 생성
  - 각비트에 해당하는 새로운 컬럼을 생성
  - 새로운 컬럼의 순서는 클래스 값을 오름차순 정렬을 한 후 에 순서대로 생성
  - 컬럼을 계속 생성 ex) A) 국어 수학 과학 
                         1    0    0
                         0    1    0
                         0    0    1  
                         
  - pandas에서 one hot encoding을 위한 함수를 제공 : get_dummies()
    - one hot encoding에 해당하는 컬럼을 생성하고 기존 컬럼은 삭제
    
  - one hot encoding의 장점
    - 새로운 특성을 늘려주는 효과 - 과소적합 감소
    - 크기를 갖지않는 범주형 데이터를 크기를 갖지 않는 수치형으로 변환

```python
import pandas as pd

data = pd.DataFrame({"산이름":["백두산", "한라산", " 무등산"]})

data
```

```python
one_hot = pd.get_dummies(data["산이름"])

one_hot
```

```python
# mushroom 데이터를 원핫인코딩
# 원핫인코딩을 한것을 다시 하면 오류가 뜬다
X2 = X.copy()

one_hot_X = pd.get_dummies(X2)

one_hot_X.head()
```

```python
# 훈련데이터와 데스트데이터로 분리
from sklearn.model_selection import train_test_split

# test_size=0.3 : 테스트데이터 비율을 30%로 설정
X_train, X_test, y_train, y_test = train_test_split(one_hot_X, y, random_state=999,
                                                    test_size=0.3)

X_train.shape, X_test.shape, y_train.shape, y_test.shape
```

# 모델 학습

```python
from sklearn.tree import DecisionTreeClassifier

model_df = DecisionTreeClassifier()
model_df.fit(X_train, y_train)
```

```python
print("훈련정확도 : ", model_df.score(X_train, y_train))
print("테스트정확도 : ", model_df.score(X_test, y_test))
```

## 사전 가지치기(pre_prunning)

- 트리의 복잡도를 감소시켜서 과대적합을 줄여주는 방법
- 종류
  - max_depth : 트리의 층을 제한
  - max_leaf_nodes : leaf node의 수를 제한
  - min_samples_split : node를 분리하는 최소 데이터의 수

### max_depth를 이용한 사전가지치기

```python
model_df2 = DecisionTreeClassifier(max_depth=3)
model_df2.fit(X_train, y_train)

print("훈련정확도 : ", model_df2.score(X_train, y_train))
print("테스트정확도 : ", model_df2.score(X_test, y_test))
```

```python
export_graphviz(model_df2, out_file="tree2.dot",
                class_names=["독", "식용"],
                feature_names= one_hot_X.columns,
                impurity=True,
                filled=True,
                rounded=True)
with open("tree2.dot", encoding="UTF8") as f:
    # 파일 읽기
    dot_graph = f.read()
    
#그래프 출력 
display(graphviz.Source(dot_graph))
```

## 교차검증(cross validation)

- 높은 점수를 찾는 것이 아닌 결과값의 평균점수를 구해 일반화를 하려고 교차검증을 한다.

```python
from sklearn.model_selection import cross_val_score
import numpy as np

# cv=5 : 데이터셋의 갯수(훈련 테스트비율 7:3인데 그 데이터셋을 가짓수가 5개 이다.)
score = cross_val_score(model_df, X_test, y_test, cv=5)
print("교차검증점수 :", score)
print("최종점수 :", np.mean(score))
```

## 정규화

- 편향된 데이터를 정규분포로 만드는 작업
- 모델의 성능을 향상

```python
data = pd.read_csv("./data/data_skew.csv")

data.head()
```

```python
X = data.iloc[:, :-1]
y = data.iloc[:, -1]
```

```python
X.iloc[:, 1].skew()
```

```python
import matplotlib.pyplot as plt

plt.xlim(0, 160)
plt.ylim(0, 70)
# X 데이터의 1번 인덱스의 값들의 개수를 저장
bins = np.bincount(X.iloc[:,1])
# x축 : 값들의 종류를 표시, y축 : 값 (개수)
plt.bar(range(len(bins)), bins)
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=555)
```

```python
from sklearn.neighbors import KNeighborsRegressor
from sklearn.metrics import mean_squared_error

knn_model = KNeighborsRegressor(n_neighbors=3)
knn_model.fit(X_train, y_train)

pred_train = knn_model.predict(X_train)
pred_test = knn_model.predict(X_test)

print("훈련오차 :", mean_squared_error(pred_train, y_train))
print("테스트오차 :", mean_squared_error(pred_test, y_test))
```

```python
# 정규분포로 만들기
# 왼쪽으로 편항되어 있으면 log()를 오른쪽으로 편향되어 있으면 exp()를 사용
# 1을 더하는 이유 : 0이되면 lg가 무한대(inf)가 되기떄문에 이를 방지하기 위한 것
X_train2 = np.log(X_train+1)
X_test2 = np.log(X_test +1)
```

```python
X_train2.iloc[:,0].skew()  # 0에가까워져서 정규화 되었다.
```

```python
plt.hist(X_train2.iloc[:, 0], bins=25)
```

```python
knn_model.fit(X_train2, y_train)

pred_train = knn_model.predict(X_train2)
pred_test = knn_model.predict(X_test2)

print("훈련오차 :", mean_squared_error(pred_train, y_train))
print("테스트오차 :", mean_squared_error(pred_test, y_test))
```
